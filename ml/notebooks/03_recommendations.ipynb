```python
import pandas as pd
import numpy as np
import joblib


features = pd.read_csv("../data/processed/features.csv")
numeric_features = features.select_dtypes(include='number')

tracks = pd.read_csv("../data/raw/tracks.csv")  # contains song name & artist

scaler = joblib.load("../models/scaler.pkl")
similarity_model = joblib.load("../models/similarity_model.pkl")

song_idx = 37152
song_features = scaler.transform([numeric_features.iloc[song_idx]])


distances, indices = similarity_model.kneighbors(song_features, n_neighbors=20)

ref_idx = 37152  # Alec Benjamin
other_idx = 123  # 2Pac - Hit Em Up

diff = numeric_features.iloc[ref_idx] - numeric_features.iloc[other_idx]
print(diff.sort_values(ascending=False))

print(f"Reference song: {tracks.iloc[song_idx]['track_name']} - {tracks.iloc[song_idx]['artists']}\n")
print("Top 20 similar songs:")

for rank, idx in enumerate(indices[0], start=1):
    song = tracks.iloc[idx]
    dist = distances[0][rank-1]
    print(f"{rank:02d}. {song['track_name']} - {song['artists']} (distance: {dist:.4f}) genre : {song['track_genre']}")
```
